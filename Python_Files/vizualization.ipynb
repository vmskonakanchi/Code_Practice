```python
import matplotlib.pyplot as pl
import numpy as np
```

```python
#g(x)= x**4-4x**2+5
```

```python
def g(x):
    return  x**4-4*x**2+5
```

```python
x_2 = np.linspace(3,-3, 1000)
def dg(x):
    return 4*x**3 - 8*x
```

```python
pl.figure(figsize=(30,10))
pl.subplot(1,2,1)
pl.plot(x_2,g(x_2),color = 'red')
pl.subplot(1,2,2)
pl.plot(x_2,dg(x_2),color= 'red')
pl.show()
```

```python
new_x = 3
previous_x = 3
gamma = 0.1
x_list = np.array(g(x_2))
slope_list  = np.array(dg(new_x))
for i in range(2):
    previous_x = new_x
    new_x = previous_x - gamma*dg(previous_x)
    x_list.__add__(new_x)
    slope_list.__add__(dg(new_x))
```

```python
pl.figure(figsize=(30,20))
pl.subplot(1,2,1)
pl.scatter(x_list,g(x_list),color = 'red')
pl.subplot(1,2,2)
pl.scatter(x_list,dg(x_list),color = 'blue')
pl.show()
```
