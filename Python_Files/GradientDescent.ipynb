```python
import numpy as np
import matplotlib.pyplot as plt

x2 = np.linspace(-2.5,2.5,1000)
def h(x):
    return x**5 - 2*x**4 + 2
     

def dh(x):
    return 5*x**4 - 8*x**3 
  

plt.figure(figsize=(30,20)) 
plt.subplot(1,2,1)
plt.plot(x2,h(x2))   
plt.subplot(1,2,2) 
plt.plot(x2,dh(x2)) 
plt.show()
  
```
