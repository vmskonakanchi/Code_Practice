```python
import pandas as pd
from matplotlib import pyplot as pp
from sklearn import linear_model as lr

data = pd.read_csv(r"C:\Users\vmsko\OneDrive\Documents\GitHub\Code_Practice\Python_Files\NUOCUT PROJECTS\cost_revenue_clean.csv")
```

```python
cost = data[["production_budget_usd"]]
returnamount = data[["worldwide_gross_usd"]]
```

```python
pp.figure(figsize=(30,30))
pp.xlabel("Production Cost in millions",fontsize = 20)
pp.ylabel("Collections in millions",fontsize = 20)
pp.scatter(cost,returnamount,color = 'red',alpha=0.1)
pp.show()
```

```python
regressionline = lr.LinearRegression()
```

```python
line = regressionline.fit(returnamount,cost)
```

```python
predtitedcost = line.predict(returnamount)
```

```python
score  = regressionline.score(returnamount,cost)
```

```python
round(score * 100,2)
```
